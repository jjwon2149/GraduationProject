```python
from glob import glob
import os

path = r'C:\Users\jaehy\Desktop\no_punch/'

file_list = glob(path + '*.txt')
arr = [0] * 8
cnt = 0
for path in file_list:
    f = open(path, 'r')
    lines = f.readlines()
    for line in lines:
        cnt += 1
        index = int(list(line.split())[0])
        arr[index] += 1
    f.close()
print(cnt)
print(arr)
```
